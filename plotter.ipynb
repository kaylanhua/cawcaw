```python


def plot_points_and_edges(points, edge, chosen_points):
    plt.figure(figsize=(8, 8))
    x, y = points[:, 0], points[:, 1]
    plt.scatter(x, y, alpha=0.6, color='blue', label='All Points')

    # Plot chosen points in a different color
    chosen_x = [points[i][0] for i in chosen_points]
    chosen_y = [points[i][1] for i in chosen_points]
    plt.scatter(chosen_x, chosen_y, alpha=1, color='red', label='Chosen Points')

    for edge in edges:
        plt.plot([x[edge[0]], x[edge[1]]], [y[edge[0]], y[edge[1]]], 'k-', alpha=0.2)

    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.title('Connected Points in a 10x10 Square (Min Distance: 0.5)')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    plt.savefig('graph.png')

plot_points_and_edges(points, edges, [7.26195424, 9.30576   ])

```
